```python
import matplotlib.pyplot as plt
```

```python
# Sample Data
y = [1, 4, 9, 16, 25]
x = [1, 2, 3, 4, 5]

# Creating the Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, marker='o', linestyle='-', color='b', linewidth=2, markersize=8, label='y = x^2')
plt.title("Basic Line Plot", fontsize=14)
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#Bar charts are useful for comparing categorical data.
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.figure(figsize=(8,5))
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
plt.title("Bar Chart Example", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Scatter plots are useful for visualizing relationships between two variables.
import numpy as np

x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.rand(50) * 300
colors = np.random.rand(50)

plt.figure(figsize=(8,5))
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.75, edgecolors='black')
plt.colorbar(label='Color Scale')
plt.title("Scatter Plot Example", fontsize=14)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#Histograms are useful for visualizing the distribution of a dataset.
data = np.random.randn(1000)

plt.figure(figsize=(8,5))
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title("Histogram Example", fontsize=14)
plt.xlabel("Data Bins", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Pie charts are useful for displaying proportions.
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 30, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0) # Highlight second slice

plt.figure(figsize=(8,5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode,
shadow=True)
plt.title("Pie Chart Example", fontsize=14)
plt.show()
```

```python
import pandas as pd

dates = pd.date_range(start="2023-01-01", periods=10)
stocks = pd.DataFrame({
'Date': dates,
'Stock Price': [150, 152, 148, 155, 160, 162, 158, 165, 168, 170]
})

plt.figure(figsize=(8,5))
plt.plot(stocks['Date'], stocks['Stock Price'], marker='o', linestyle='-', color='blue', linewidth=2)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Stock Price (USD)", fontsize=12)
plt.title("Stock Price Trends", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=sns.load_dataset('tips'))
plt.title("Total Bill vs Tip Amount")
plt.show()
```

```python
#Line Plot (Time Series Trends) Used for tracking trends over continuous data points.
sns.lineplot(x='size', y='total_bill', data=sns.load_dataset('tips'))
plt.title("Meal Size vs. Total Bill")
plt.show()
```

```python
#Bar Plot (Comparing Categories) compares gender-based differences in spending.
sns.barplot(x='day', y='total_bill', hue='sex', data=sns.load_dataset('tips'))
plt.title("Total Bill by Day and Gender")
plt.show()
```

```python
sns.countplot(x='day', hue='sex', data=sns.load_dataset('tips')) #Shows how many entries exist for each category.
plt.title("Number of Customers per Day")
plt.show()
```

```python
#Box Plot (Identifying Outliers and Distribution) Displays median, quartiles, and outliers.
sns.boxplot(x='day', y='total_bill', hue='sex', data=sns.load_dataset('tips'))
plt.title("Total Bill Distribution by Day and Gender")
plt.show()
```
